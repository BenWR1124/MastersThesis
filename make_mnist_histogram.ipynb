```python
import numpy as np
from matplotlib import pyplot as plt
from keras.datasets import mnist

# Import the MNIST digits
(train_X, train_y), (test_X, test_y) = mnist.load_data()
train_X = train_X.reshape(train_X.shape[0], 28*28)
test_X = test_X.reshape(test_X.shape[0], 28*28)


# Set the file format and name
image_format = 'png'
image_name = 'mnistHist.png'

# Make the histogram
fig, ax = plt.subplots()
plt.hist(train_y)
plt.xticks(ticks=np.arange(0, 10, step=1),labels=range(10))
plt.xlabel('Digit')
plt.ylabel('Quantity')
plt.hist(test_y)
plt.legend(['Training Data','Test Data'],loc=7)
plt.title('MNIST Digit Distrubution')
fig.savefig(image_name, format=image_format, dpi=1200)
```
