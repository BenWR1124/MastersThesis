```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Create the data
rc=[53.5,56.0,51.7,51.9,52.3]
cs=[55.5,61.0,52.3,53.3,55.0]
```

```python
# Plot Score Comparison
plt.plot(range(5),cs)
plt.plot(range(5),rc)
plt.xticks(ticks=range(5),labels=["S1","S2","S3","S4","S5"])
plt.legend(["Compressive Sensing","Reverse Correlation"])
plt.xlabel("Subject")
plt.ylabel("Accuracy (%)")
plt.title("Balanced Accuracies from Cross-Validation")
plt.show()
```

```python
 # Make Bar Plots Comparing Scores
plt.bar(np.linspace(0.83,4.83,5),cs,width=0.325,color='black')
plt.bar(np.linspace(1.17,5.17,5),rc,width=0.325,color='gray')
plt.ylim([48,63])
plt.legend(["Compressive Sensing","Reverse Correlation"])
plt.ylabel("Accuracy (%)")
plt.xticks(ticks=range(1,6),labels=["S1","S2","S3","S4","S5"])
plt.show()
```
